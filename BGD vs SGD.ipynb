```python
epochs=2500   #no.of interation
m=0
b=0
lr=0.001
batch_size=X_train.shape[0]

for i in range(epochs):
    
    y_pred=m * X_train.reshape(77) + b   #y=mx+b
    
    loss= sum(y_train - y_pred)** 2/len(X_train)
    
    d_m=(-2/len(X_train)) * sum((y_train - y_pred) * X_train.reshape(77))
    
    d_b=(-2/len(X_train)) * sum(y_train - y_pred)
    
    m=m-(lr*d_m)
    b=b-(lr*d_b)
    
    print("The loss after",i,"iteration",loss)
```
